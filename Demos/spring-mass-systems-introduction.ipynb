```python
# load the necessary packages
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# define the function that returns dz/dt
def myModel(y, t):
    #
    # parameters
    f = 1.1
    k = 0.4
    m = 0.4
    #
    # get the individual variables - for readability
    x0 = y[0]
    x1 = y[1]
    #
    # define a step input
    if t > 10:
        u = 1
    else:
        u = 0
    #
    # individual derivatives
    dx0dt  = x1
    dx1dt  = - (f / m) * x1 - (k / m) * x0 + (1 / m) * u
    #
    return [ dx0dt, dx1dt ]

# define the initial condition
y0 = [ 2, 5 ]

# define the time points where the solution is computed
n    = 100
tmax = 20
t    = np.linspace(0, tmax, n)

# solve the ODE
y = odeint(myModel, y0, t)

# get the individual variables
x0 = y[:,0]
x1 = y[:,1]

# plot the time evolution
plt.subplot(1,2,1)
plt.plot(t, x0, label='position', linewidth=1)
plt.plot(t, x1, label='velocity', linewidth=1)
plt.grid()
plt.legend(loc = 'upper right')
plt.xlabel('time evolution')

# plot the phase portrait
plt.subplot(1,2,2)
plt.plot(x0, x1, linewidth=1)
plt.grid()
plt.xlabel('position')
plt.ylabel('velocity')
plt.title('phase portrait')

plt.show()

```
